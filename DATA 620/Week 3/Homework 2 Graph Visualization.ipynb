## Adding in Characters

Now that we have all the movies in place, we can add in the all the characters. To add new characters, we can use the syntax defined in the first secion, and to add in an existing characters such as George Clooney, we can use the following Cypher syntax:

```python
MATCH (u:Actor {nickname:'clooney'}), (r:Movie {name:"Ocean's 12"})
CREATE (u)-[:in]->(r)
```

## Measuring Diameter

To measure the diameter, the following piece of code was used. It looks for all connections between two nodes, and returns the total count in decescending order. It was found via this [StackOverflow thread](https://stackoverflow.com/questions/18072250/node-degree-query-on-neo4j-with-huge-graph-million-nodes-and-links).

```python
START n = node(*) 
MATCH (n)--(c)
RETURN n, count(*) as connections
ORDER BY connections DESC
LIMIT 10
```

The code above generates the following graph:

```python
from IPython.display import Image
Image(filename='graph_9.png') 
```

As we can see, the longest distance is when we start from either George Clooney, Bernie Mac or Eddie Jemison, and end up at Matt Damon or Carl Renner. This diameter spans all four movies, thus the size of the diameter is four. 

## Calculating Number of Appearences

Lastly, I calculated the number of appearences in the movies by actor using the following Cypher code:

```python
MATCH (actor:Actor)-[:in]->(movie:Movie) 
RETURN actor.name, COUNT(actor) AS actor
ORDER BY actor DESC
```
